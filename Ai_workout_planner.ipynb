```python
import random

print("\n===========================================")
print("üí™ AI PERSONALIZED WORKOUT & DIET PLANNER üí™")
print("===========================================\n")

# -------------------------------
# Collect user information
# -------------------------------
name = input("Enter your name: ").capitalize()
age = int(input("Enter your age: "))
gender = input("Enter your gender (Male/Female): ").lower()
weight = float(input("Enter your weight (kg): "))
height = float(input("Enter your height (cm): "))
activity = input("Activity level (low / medium / high): ").lower()
goal = input("Goal (gain / lose / maintain): ").lower()
diet_type = input("Diet preference (veg / nonveg): ").lower()

# -------------------------------
# Calculate BMR & Calories
# -------------------------------
if gender == "male":
    bmr = 10 * weight + 6.25 * height - 5 * age + 5
else:
    bmr = 10 * weight + 6.25 * height - 5 * age - 161

# Adjust for activity
if activity == "low":
    calories = bmr * 1.2
elif activity == "medium":
    calories = bmr * 1.55
else:
    calories = bmr * 1.75

# Adjust for goal
if goal == "gain":
    calories += 300
elif goal == "lose":
    calories -= 300

# -------------------------------
# Workout Plan Generator
# -------------------------------
def workout_plan(goal):
    if goal == "gain":
        return {
            "Monday": "Chest & Triceps (Bench Press:incline,decline,flat Push-ups, Dips,triceps cable Extensions:Neutral grip,revrse grip,overhead dumbell,cable cross done)",
            "Tuesday": "Back & Biceps (Pull-ups, Rows, deadlift,lat pulldowns,dumbell Curls,hammer crules,burbell crules,preacher curls)",
            "Wednesday": "Rest or Cardio",
            "Thursday": "Legs (Squats, Lunges, Leg Press,Leg extensions)",
            "Friday": "Shoulders (Overhead Press, Lateral Raises,arnold press,front raises,reardelts rasies)",
            "Saturday": "Core (Plank, Crunches, Leg Raises)",
            "Sunday": "Rest"
        }
    elif goal == "lose":
        return {
            "Monday": "Full Body (Burpees, Squats, Jumping Jacks)",
            "Tuesday": "Cardio (Running, Cycling, Jump Rope)",
            "Wednesday": "Core (Planks, Crunches, Mountain Climbers)",
            "Thursday": "Legs + Cardio Mix",
            "Friday": "Arms & Shoulders (Light weights or push-ups)",
            "Saturday": "Yoga or Stretching",
            "Sunday": "Rest"
        }
    else:
        return {
            "Monday": "Push (Chest, Shoulders, Triceps)",
            "Tuesday": "Pull (Back, Biceps)",
            "Wednesday": "Rest",
            "Thursday": "Legs",
            "Friday": "Core + Cardio",
            "Saturday": "Fun activity or sports",
            "Sunday": "Rest"
        }

# -------------------------------
# Diet Plan Generator
# -------------------------------
veg_foods = [
    "Oats with milk and banana",
    "Paneer rice bowl",
    "Lentils (Dal) with brown rice",
    "Chapati with mixed vegetables",
    "Curd and salad",
    "Fruit smoothie",
]

nonveg_foods = [
    "Boiled eggs with toast",
    "Chicken breast with rice",
    "Fish curry with chapati",
    "Boiled eggs",
    "Curd with chicken salad",
    "Protein shake with banana",
]

def meal_plan(diet, calories):
    if diet == "veg":
        foods = random.sample(veg_foods, 6)
    else:
        foods = random.sample(nonveg_foods, 6)
    print("\n--- Sample Daily Diet Plan ---")
    print(f"Target Calories: {int(calories)} kcal/day")
    for i, food in enumerate(foods, 1):
        print(f"Meal {i}: {food}")

# -------------------------------
# Display Result
# -------------------------------
print("\n===========================================")
print(f"Hello {name} üí™ Here‚Äôs your Personalized Plan!")
print("===========================================\n")

print(f"Estimated BMR: {int(bmr)} kcal/day")
print(f"Recommended Daily Calories: {int(calories)} kcal/day\n")

print("üèãÔ∏è Workout Plan:")
for day, routine in workout_plan(goal).items():
    print(f"{day}: {routine}")

meal_plan(diet_type, calories)

print("\nüí° Tip: Track your progress every 2 weeks and adjust your calories by ¬±200 kcal if needed!")
print("===========================================\n")
```
